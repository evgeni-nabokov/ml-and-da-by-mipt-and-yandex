```python
%matplotlib inline

import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as pp

x_points = np.arange(1, 16, 0.5)

# Целевая функция.
def f(x):
    return np.sin(x / 5.0) * np.exp(x / 10.0) + 5 * np.exp(-x / 2.0)

pp.plot(x_points, f(x_points), "-r")



# Линейное приближение.
x = np.array([1, 15]).reshape(-1, 1)
b = f(x).reshape(-1, 1)
A = np.array([[1, 1], [1, 15]])
w = la.solve(A, b)

def f1(x):
    return w[0, 0] + w[1, 0] * x

pp.plot(x_points, f1(x_points), "-b")




# Квадратичное приближение.
x = np.array([1, 8, 15]).reshape(-1, 1)
b = f(x).reshape(-1, 1)
A = np.array([[1, 1, 1], [1, 8, 8**2], [1, 15, 15**2]])
w = la.solve(A, b)

def f2(x):
    return w[0, 0] + w[1, 0] * x + w[2, 0] * (x**2)

pp.plot(x_points, f2(x_points), "-c")



# Кубическое приближение.
x = np.array([1, 4, 10, 15]).reshape(-1, 1)
b = f(x).reshape(-1, 1)
A = np.array([[1, 1, 1, 1], [1, 4, 4**2, 4**3], [1, 10, 10**2, 10**3], [1, 15, 15**2, 15**3]])
w = la.solve(A, b)

def f3(x):
    return w[0, 0] + w[1, 0] * x + w[2, 0] * (x**2) + w[3, 0] * (x**3)

pp.plot(x_points, f3(x_points), "-m")

# Сохранение значений приближённой функции в 4-х точках.
file_obj = open('hw22.txt', 'w')
answer = " ".join([format(i[0]) for i in w])
file_obj.write(answer)
file_obj.close()

# Рисование всех графиков.
pp.show()
```
